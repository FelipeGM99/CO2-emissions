```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the data from csv frame to Pandas DataFrame
CO2_dataset = pd.read_csv('co-emissions-by-sector.csv')
```

```python
CO2_dataset.head()
```

```python
df_filtered_EUA = CO2_dataset[(CO2_dataset['Entity'] == 'United States') & (CO2_dataset['Year'] >= CO2_dataset['Year'].max() - 9)]
df_filtered_UE = CO2_dataset[(CO2_dataset['Entity'] == 'European Union (27)') & (CO2_dataset['Year'] >= CO2_dataset['Year'].max() - 9)]
df_filtered_MEX = CO2_dataset[(CO2_dataset['Entity'] == 'Mexico') & (CO2_dataset['Year'] >= CO2_dataset['Year'].max() - 9)]
```

```python
plt.plot(df_filtered_EUA['Year'], df_filtered_EUA['Carbon dioxide emissions from transport'], label='EUA')
plt.plot(df_filtered_UE['Year'], df_filtered_UE['Carbon dioxide emissions from transport'], label='UE (27)')
plt.plot(df_filtered_MEX['Year'], df_filtered_MEX['Carbon dioxide emissions from transport'], label='Mexico')
plt.xlabel('Año')
plt.ylabel('Emisiones de CO2 causadas por transporte (t)')
plt.legend()
plt.show()

dataframes = [df_filtered_EUA, df_filtered_UE, df_filtered_MEX]
nombres_paises = ['Estados Unidos', 'Unión Europea', 'México']

for pais, nombre in zip(dataframes, nombres_paises):
  year_max_emissions = pais.loc[pais['Carbon dioxide emissions from transport'].idxmax(), 'Year']
  max_emissions_value = pais['Carbon dioxide emissions from transport'].max()
  print(f"El año con las emisiones máximas de CO2 en {nombre} causadas por transporte fue {year_max_emissions} con {max_emissions_value} toneladas.")


```

```python
plt.plot(df_filtered_EUA['Year'], df_filtered_EUA['Carbon dioxide emissions from electricity and heat'], label='EUA')
plt.plot(df_filtered_UE['Year'], df_filtered_UE['Carbon dioxide emissions from electricity and heat'], label='UE (27)')
plt.plot(df_filtered_MEX['Year'], df_filtered_MEX['Carbon dioxide emissions from electricity and heat'], label='Mexico')
plt.xlabel('Año')
plt.ylabel('Emisiones de CO2 causadas por \n la industria eléctrica (t)')
plt.legend()
plt.show()

dataframes = [df_filtered_EUA, df_filtered_UE, df_filtered_MEX]
nombres_paises = ['Estados Unidos', 'Unión Europea', 'México']

for pais, nombre in zip(dataframes, nombres_paises):
  year_max_emissions = pais.loc[pais['Carbon dioxide emissions from electricity and heat'].idxmax(), 'Year']
  max_emissions_value = pais['Carbon dioxide emissions from electricity and heat'].max()
  print(f"El año con las emisiones máximas de CO2 en {nombre} causadas por la industria eléctrica fue {year_max_emissions} con {max_emissions_value} toneladas.")
```

```python
plt.plot(df_filtered_EUA['Year'], df_filtered_EUA['Carbon dioxide emissions from manufacturing and construction'], label='EUA')
plt.plot(df_filtered_UE['Year'], df_filtered_UE['Carbon dioxide emissions from manufacturing and construction'], label='UE (27)')
plt.plot(df_filtered_MEX['Year'], df_filtered_MEX['Carbon dioxide emissions from manufacturing and construction'], label='Mexico')
plt.xlabel('Año')
plt.ylabel('Emisiones de CO2 causadas por \n la industria manufacturera y \n de construcción (t)')
plt.legend()
plt.show()

dataframes = [df_filtered_EUA, df_filtered_UE, df_filtered_MEX]
nombres_paises = ['Estados Unidos', 'Unión Europea', 'México']

for pais, nombre in zip(dataframes, nombres_paises):
  year_max_emissions = pais.loc[pais['Carbon dioxide emissions from manufacturing and construction'].idxmax(), 'Year']
  max_emissions_value = pais['Carbon dioxide emissions from manufacturing and construction'].max()
  print(f"El año con las emisiones máximas de CO2 en {nombre} causadas por la industria manufacturera y de construcción fue {year_max_emissions} con {max_emissions_value} toneladas.")
```
